```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
file1=pd.read_csv('diabetes.csv')
file2=pd.read_excel('avocado.xls')
```

```python
file1.head()
```

```python
file2.head()
```

```python
x=file1['BMI']
y=file1['Glucose']
plt.figure(figsize=(15,7))
plt.plot(x,y,linewidth=2,marker='*',linestyle='dashed',color='green')
plt.xlabel('BMI')
plt.ylabel('Glucose')
plt.title('plotting')
plt.legend()
```

```python
plt.figure(figsize=(15,7))
plt.plot('Small Bags','Large Bags','go-',data=file2,label='line1')
plt.plot('Small Bags','Total Bags','rs',data=file2,label='line2')
plt.legend()
```

```python
plt.figure()
plt.plot('Small Bags','Large Bags',data=file2)
plt.figure()
plt.plot('Small Bags','Total Bags',data=file2)
```

```python
x=plt.figure()
y=plt.figure()
sub=x.add_axes([0.1,0.1,0.5,0.5])
sub1=y.add_axes([0.1,0.1,0.8,0.8])
sub.plot('Small Bags','Large Bags',data=file2)
sub1.plot('Small Bags','Total Bags',data=file2)
```
